```python
# Preparing the data
df_prophet = pd.DataFrame(df.resample('M').size().reset_index())
```

```python
df_prophet.columns = ['Date','Crime Count']
```

```python
df_prophet
```

```python
# Rename the columns
# ds and y
```

```python
df_prophet =df_prophet.rename(columns={'Date':'ds','Crime Count':'y'})
```

```python
df_prophet
```

```python
#Train the model
m = Prophet()
m.fit(df_prophet)
```

```python
#Make the Predictions
pred = m.make_future_dataframe(periods=24,freq='M')
forcast = m.predict(pred)
```

```python
forcast
```

```python
m.plot(forcast,xlabel='Data',ylabel='Crime Rate')
```

```python
# save the model
import json
from prophet.serialize import model_to_json,model_from_json
with open('prophet_model.json','w') as fout:
    json.dump(model_to_json(m),fout) #save the model
```
